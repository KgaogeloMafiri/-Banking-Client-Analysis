```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
# ── STYLE SETTINGS ───────────────────────────────────────────────────────────
plt.rcParams['figure.facecolor']  = '#F8FAFC'
plt.rcParams['axes.facecolor']    = '#FFFFFF'
plt.rcParams['axes.edgecolor']    = '#C5D8E8'
plt.rcParams['axes.labelcolor']   = '#0C1B33'
plt.rcParams['axes.titlecolor']   = '#0C1B33'
plt.rcParams['axes.titleweight']  = 'bold'
plt.rcParams['axes.titlesize']    = 13
plt.rcParams['xtick.color']       = '#555555'
plt.rcParams['ytick.color']       = '#555555'
plt.rcParams['font.family']       = 'DejaVu Sans'
plt.rcParams['grid.color']        = '#E2E8F0'
plt.rcParams['grid.linestyle']    = '--'
plt.rcParams['grid.alpha']        = 0.6

NAVY  = '#0C1B33'
TEAL  = '#0EA5A0'
GOLD  = '#F59E0B'
RED   = '#E53E3E'
GRAY  = '#64748B'
PALETTE = [TEAL, NAVY, GOLD, RED, '#7C3AED', '#059669', '#DB2777', '#D97706']
```

# Data loading, column overview, data types, null counts, duplicates


```python
print("=" * 60)
print("  BANKING DATASET — EXPLORATORY DATA ANALYSIS")
print("=" * 60)

df = pd.read_csv('/Banking.csv',
                 parse_dates=['Joined Bank'], dayfirst=True)

print(f"\n✔ Dataset loaded successfully")
print(f"  Rows    : {df.shape[0]:,}")
print(f"  Columns : {df.shape[1]}")

# ── Column overview ──────────────────────────────────────────
print("\n── COLUMN OVERVIEW ─────────────────────────────────────")
col_info = pd.DataFrame({
    'Data Type'   : df.dtypes,
    'Non-Null'    : df.notnull().sum(),
    'Null Count'  : df.isnull().sum(),
    'Null %'      : (df.isnull().sum() / len(df) * 100).round(2),
    'Unique Values': df.nunique()
})
print(col_info.to_string())

# ── Statistical summary ──────────────────────────────────────
print("\n── DESCRIPTIVE STATISTICS (Numeric Columns) ────────────")
numeric_cols = df.select_dtypes(include='number').columns.tolist()
print(df[numeric_cols].describe().round(2).to_string())

# ── Categorical summary ──────────────────────────────────────
print("\n── CATEGORICAL COLUMNS ─────────────────────────────────")
cat_cols = df.select_dtypes(include='object').columns.tolist()
for col in cat_cols:
    print(f"\n  {col} ({df[col].nunique()} unique values):")
    print(df[col].value_counts().head(5).to_string())

# ── Duplicate check ──────────────────────────────────────────
dupes = df.duplicated(subset='Client ID').sum()
print(f"\n── DUPLICATES ──────────────────────────────────────────")
print(f"  Duplicate Client IDs: {dupes}")
```

# Missing values heatmap


```python
fig, ax = plt.subplots(figsize=(14, 4))
missing = df.isnull()
sns.heatmap(missing, cbar=False, cmap=['#FFFFFF', RED],
            yticklabels=False, ax=ax, linewidths=0.1)
ax.set_title('Missing Values Heatmap  (Red = Missing)', pad=12)
ax.set_xlabel('Columns')
plt.tight_layout()
plt.savefig('01_missing_values.png', dpi=150, bbox_inches='tight')
plt.show()
print("\n✔ Plot saved: 01_missing_values.png")
```

# Correlation heatmap of all numeric variables

```python
fig, ax = plt.subplots(figsize=(14, 10))
corr = df[numeric_cols].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(
    corr, mask=mask, annot=True, fmt='.2f',
    cmap=sns.diverging_palette(220, 20, as_cmap=True),
    center=0, square=True, linewidths=0.5,
    annot_kws={'size': 8}, ax=ax,
    cbar_kws={'shrink': 0.8}
)
ax.set_title('Correlation Matrix — Numeric Variables',
             fontsize=14, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig('07_correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
print("✔ Plot saved: 07_correlation_heatmap.png")
```

# Outlier detection using box plots with IQR method

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()
fig.suptitle('Outlier Detection — Box Plots', fontsize=15,
             fontweight='bold', color=NAVY)

outlier_cols = ['Estimated Income', 'Bank Loans', 'Bank Deposits',
                'Credit Card Balance', 'Business Lending', 'Superannuation Savings']

for i, col in enumerate(outlier_cols):
    ax = axes[i]
    bp = ax.boxplot(df[col], patch_artist=True, vert=True,
                    medianprops={'color': GOLD, 'linewidth': 2},
                    flierprops={'marker': 'o', 'color': RED,
                                'alpha': 0.4, 'markersize': 4})
    bp['boxes'][0].set_facecolor(TEAL)
    bp['boxes'][0].set_alpha(0.6)

    # Calculate outlier count
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
    ax.set_title(f'{col}\n({len(outliers)} outliers detected)')
    ax.set_ylabel('Value')
    ax.yaxis.set_major_formatter(mticker.FuncFormatter(
        lambda x, _: f'{x/1e6:.1f}M' if x >= 1e6 else f'{x/1e3:.0f}K'))
    ax.grid(True, alpha=0.4, axis='y')

plt.tight_layout()
plt.savefig('09_outliers.png', dpi=150, bbox_inches='tight')
plt.show()
print("✔ Plot saved: 09_outliers.png")
```
